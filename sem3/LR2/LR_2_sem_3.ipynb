Вопрос - как нам решить обратную задачу? Т.е., имея размытое изображение, как сделать его более четким? Попробуйте подобрать матрицу, которая может это сделать.

```python
img = cv.imread("./starry_night.jpg")
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)

plt.imshow(img)
```

# Задание 1:

```python
def increase_brightness(img, value=30):
    hsv = cv.cvtColor(img, cv.COLOR_RGB2HSV)
    h, s, v = cv.split(hsv)
    lim = 255 - value
    v[v > lim] = 255
    v[v <= lim] += value
    final_hsv = cv.merge((h, s, v))
    return cv.cvtColor(final_hsv, cv.COLOR_HSV2RGB)
```

```python
new_img = increase_brightness(img, 80)
```

```python
plt.imshow(new_img)
```

```python
img_blur = cv.blur(img, (7, 7))
```

```python
plt.imshow(img_blur)
```

Ядро для востановления размытых изображений

```python
kernel = np.array([
    [0, -1, 0],
    [-1, 5, -1],
    [0, -1, 0]
])
```

```python
filtered_img = cv.filter2D(img_blur, -1, kernel)
```
